Задание 2
Напишите код на Python для решения следующей задачи.
Реализуйте функцию trim_and_repeat(), которая принимает три параметра:
● строку;
● offset — число символов, на которое нужно обрезать строку слева;
● repetitions — сколько раз нужно повторить строку перед возвратом
получившейся строки.
Число символов для среза по умолчанию равно 0, а количество повторений —
1.
Функция должна возвращать полученную строку

```python
string = input('Введите строку:')
offset = int(input('Укажите какое число символов нужно обрезать от строки, начиная с ее левой части:'))
repetitions = int(input('Укажите сколько раз нужно повторить строку перед возвратом получившейся строки:'))

cut_string = string[offset:]*repetitions #используем обрезание, затем повторяем строку заданное количество раз

print(cut_string)
```
